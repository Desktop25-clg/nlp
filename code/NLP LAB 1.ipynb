```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "NLP is study of proceesing of langugae by AI."

doc = nlp(text)

tokens = [token.text for token in doc ]
lemma = [token.lemma_ for token in doc ]

print(tokens)
print(lemma)

print("Dependecy Prasing\n")

for token in doc:
  print(token.text,token.dep_,token.head,token.head.text,token.head.pos_,[child for child in token.children])
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

customer_feedback = [
"The product is amazing! I love the quality.",
"The customer service was terrible, very disappointed.",
"Great experience overall, highly recommended.",
"The delivery was late, very frustrating."]

def analyze_feedback(feedback):
  for idx,text in enumerate(feedback,start=1):
    doc = nlp(text)
    print("\n")
    print(f"Analyzing feedback {idx} : ,'{feedback}'")

    tokens = [token.text for token in doc]
    lemma = [token.lemma_ for token in doc]

    print(f"tokens : {tokens}")
    print(f"lemma : {lemma}")

    print("Depency Prasing: \n")
    for token in doc:
      print(token.text,token.dep_,token.head.text,token.pos_,[child for child in token.children])
  print("\n")
if __name__ == "__main__":
  analyze_feedback(customer_feedback)

```
